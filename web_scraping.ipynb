## 1: Scraping using Pandas libraries


```python
import pandas as pd
import requests
from io import StringIO  # ← Yeh add karo

url = "https://en.wikipedia.org/wiki/Demographics_of_Pakistan"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# Step 1: HTML fetch karo
response = requests.get(url, headers=headers)

# Step 2: StringIO mein wrap karo ← Yahi fix hai
tables = pd.read_html(StringIO(response.text))

# Step 3: Tables check karo
print(f"Number of tables found: {len(tables)}")

# Step 4: Pehli table dekho
tables[4].head()  # ← Yeh change karo, 23 se pehle check karo ki kitni tables hain
```

```python
import pandas as pd

url = "https://www.worldometers.info/world-population/population-by-country/"

# Fetch the HTML content
import requests

# header to avoid blocking by the website
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}

response = requests.get(url, headers=headers)
response.raise_for_status()

from io import StringIO
tables = pd.read_html(StringIO(response.text))
print(f"Number of tables found: {len(tables)}")

tables[0].head()  # First table is usually the one we want

# # sAVE THE DATA TO CSV
# tables[0].to_csv("population_by_country_2-6-2026.csv", index=False)
```
